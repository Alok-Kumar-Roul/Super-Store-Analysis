# Conclusion :-
- In summary, the dataset analysis highlights significant patterns. Notably, "Ship Mode" favors the Standard class (3451) over Same day (344).
```python
ax = sns.catplot(x="Order Month", y="Sales", data=df, kind="bar", estimator=sum, height=6, aspect=2)
plt.subplots_adjust(top=0.9)
ax.fig.suptitle("Total Sales by Month") 

for p in ax.ax.patches:
    ax.ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=8)

plt.show()
```

```python
ax = sns.catplot(x="Order Year", y="Sales", data=df, kind="bar", estimator=sum)
plt.subplots_adjust(top=0.9)
ax.fig.suptitle("Total Sales by Month") 

for p in ax.ax.patches:
    ax.ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=8)

plt.show()
```
